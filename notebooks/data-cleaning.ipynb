```python
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
```

```python
with open('../data/Sleep/sleep-2016-04-02.json') as json_file:
    data = json.load(json_file)
```

```python
data_df = pd.DataFrame(data)
data_df
```

```python
sleep_summary = pd.json_normalize(data_df['levels'])
sleep_summary
```

```python
# sleep_summary.drop('data',axis=1,inplace=True)

data_df = pd.concat([data_df,sleep_summary], axis=1)

data_df.columns
```
